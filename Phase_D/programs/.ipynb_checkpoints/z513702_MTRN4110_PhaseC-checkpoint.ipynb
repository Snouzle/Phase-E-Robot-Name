```python
#import libraries 
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

kernel = np.ones((3,3), np.uint8)

# lower and upper ranges for pink
lower_pink = np.array([140,166,243])
upper_pink = np.array([156,255,255])


#lower and upper ranges for cyan
lower_cyan = np.array([87,90,90])
upper_cyan = np.array([92,255,255])

mask_pink = cv2.inRange(img_hsv, lower_pink, upper_pink)
mask_cyan = cv2.inRange(img_hsv, lower_cyan, upper_cyan)

mask_pink = cv2.medianBlur(mask_pink,5)
mask_cyan = cv2.medianBlur(mask_cyan,5)

se1 = cv2.getStructuringElement(cv2.MORPH_RECT, (11,11))
mask_pink = cv2.morphologyEx(mask_pink, cv2.MORPH_CLOSE, se1)
mask_pink = cv2.morphologyEx(mask_pink, cv2.MORPH_CLOSE, se1)

mask_cyan = cv2.morphologyEx(mask_cyan, cv2.MORPH_CLOSE, se1)
mask_cyan = cv2.morphologyEx(mask_cyan, cv2.MORPH_CLOSE, se1)

img_pink, contours_pink, hierarchy = cv2.findContours(mask_pink, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
img_cyan, contours_cyan, hierarchy = cv2.findContours(mask_cyan, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

largestValue = len(contours_cyan[0])
cyan_index = 0

if (len(contours_cyan) > 1) :
    for x in range(len(contours_cyan)) :
        if len(contours_cyan[x]) > largestValue :
            largestValue = len(contours_cyan[x])
            cyan_index = x

cv2.drawContours(img, contours_pink,  -1, (255,0,0), 2)
cv2.drawContours(img, [contours_cyan[cyan_index]],  0, (0,0,255), 2)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)

plt.show()
```

## 3.4

```python
#import libraries 
import cv2
import numpy as np
import matplotlib.pyplot as plt
verticalMatrix = np.zeros((5,10))
row_counter = 0
col_counter = 0
horizontalMatrix = np.zeros((6,9))

img = cv2.cvtColor(dst, cv2.COLOR_RGB2BGR)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

kernel = np.ones((3,3), np.uint8)

lower = np.array([18,35,210]) #([18,35,210])
upper = np.array([32,255,255]) #([32,255,255])

mask = cv2.inRange(img_hsv, lower, upper)

img_walls, contours_wall, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

img_walls = cv2.erode(img_walls, kernel, iterations=1)
img_walls = cv2.medianBlur(img_walls, 5)
img_walls = cv2.dilate(img_walls, kernel, iterations=2)
img_walls, contours_wall, heirarhcy = cv2.findContours(img_walls, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# FINDING VERTICAL AND HORIZONTAL MATRIX
row,cols = img_walls.shape
# finding vertical wall matrix 
for i in range(50,row,100) : #checking the middle of the vertical wall
    row_counter = int(i/100)
    verticalMatrix[row_counter][0] = 1
    verticalMatrix[row_counter][9] = 1
    for j in range(100,cols,100) : #iterating through each column 
        col_counter = int(j/100)
        if img_walls[i,j] == 255 :
            verticalMatrix[row_counter][col_counter] = 1
            
# finding horizontal wall matrix 
for i in range(0,row,100) :
    row_counter = int(i/100)
    #horizontalMatrix[col_counter][0]
    for j in range(50,cols,100) :
        col_counter = int(j/100)
        if row_counter == 0 :
            horizontalMatrix[row_counter][col_counter] = 1
        elif img_walls[i,j] == 255 :
            horizontalMatrix[row_counter][col_counter] = 1
for i in range(9) :
    horizontalMatrix[5][i] = 1

# drawing vertical rectangles     
coordinate_upper_corner = (int(0),int(0))
coordinate_bottom_corner = (int(0),int(0))
for i in range(5) :
    for j in range(1,9,1) :
        if verticalMatrix[i][j] == 1 :
            bottom_corner_x = int(j*100-2.5)
            bottom_corner_y = int((i+1)*100+2.5)
            
            upper_corner_x = int(j*100+2.5)
            upper_corner_y = int(i*100)
            
            coordinate_upper_corner = (upper_corner_x, upper_corner_y)
            coordinate_bottom_corner = (bottom_corner_x,bottom_corner_y)
            cv2.rectangle(dst, coordinate_bottom_corner, coordinate_upper_corner, (0,0,255), 3)

#drawing horizontal rectangles
coordinate_upper_corner = (int(0),int(0))
coordinate_bottom_corner = (int(0),int(0))
for i in range(1,5,1) :
    for j in range(0,9,1) :
        if horizontalMatrix[i][j] == 1 :
            bottom_corner_x = int(j*100)
            bottom_corner_y = int(i*100+2.5)
            
            upper_corner_x = int((j+1)*100)
            upper_corner_y = int(i*100-2.5)
            
            coordinate_upper_corner = (upper_corner_x, upper_corner_y)
            coordinate_bottom_corner = (bottom_corner_x,bottom_corner_y)
            cv2.rectangle(dst, coordinate_bottom_corner, coordinate_upper_corner, (0,0,255), 3)
            
            
cv2.drawContours(img, contours_wall,  -1, (255,0,0), cv2.FILLED)
img_rgb_wall = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(dst)
plt.show()




```

```python
#import required libraries
import cv2 # OpenCV library
import numpy as np # Numpy library for scientific computing
import matplotlib.pyplot as plt # Matplotlib library for plotting

# Generate an ArUco marker for use
# Load the predefined dictionary

kernel = np.ones((3,3), np.uint8)
dictionary = cv2.aruco.Dictionary_get(cv2.aruco.DICT_6X6_250)

# Detect ArUco markers
# image is loaded with imread command 
# TODO: USE DEFINTION OF ROBOT FROM 3.1
img_aruco_bgr = cv2.imread(ROBOT_FILE_NAME)
img_aruco_rgb_origin = cv2.cvtColor(img_aruco_bgr,cv2.COLOR_BGR2RGB)
img_aruco_rgb = cv2.cvtColor(img_aruco_bgr,cv2.COLOR_BGR2RGB)

img_wall_epuck = cv2.cvtColor(dst, cv2.COLOR_RGB2BGR);

#img of map in HSV (without walls)
img_hsv = cv2.cvtColor(dst, cv2.COLOR_RGB2HSV)

# Initialize the detector parameters using default values
parameters =  cv2.aruco.DetectorParameters_create()

# Detect the markers in the image
markerCorners, markerIds, rejectedCandidates = cv2.aruco.detectMarkers(img_aruco_rgb, dictionary, parameters=parameters)

markerCorners_list = markerCorners[0][0].tolist()
cornersX = [];
cornersY = [];

for i in range(len(markerCorners_list)) :
    cornersX.append(markerCorners_list[i][0])
    cornersY.append(markerCorners_list[i][1])

indexX = [0]*len(cornersX)
indexY = [0]*len(cornersY)
for i in range(len(cornersX)):
    indexX[cornersX.index(min(cornersX))] = i
    indexY[cornersY.index(min(cornersY))]  = i
    
    cornersX[cornersX.index(min(cornersX))] = max(cornersX)+1
    cornersY[cornersY.index(min(cornersY))] = max(cornersY)+1
    
if indexX[0] == 3 :
    if center_list[0][0] < 337.5 :
        direction = [False, False, True, False]
        char_direction = '>'
    else :
        direction = [False, False, False, True]
        char_direction = '<'
    
elif indexX[0] == 0 :
     #west is true    
    if center_list[0][0] < 337.5 :
        direction = [False, False, False, True]
        char_direction = '<'
    else :
        direction = [False, False, True, False]
        char_direction = '>'
    
elif indexY[0] == 0 :
    #south is true 
    if center_list[0][0] < 337.5 :
        direction = [False, True, False, False]
        char_direction = '^'
    else :
        direction = [True, False, False, False]
        char_direction = 'v'
    
elif indexY[0] == 3 :
    #north is true
    if center_list[0][0] < 337.5 :
        direction = [True, False, False, False]
        char_direction = 'v'
    else :
        diection = [False, True, False, False]
        char_direction = '^'


lower_epuck = np.array([27,50,126])
upper_epuck = np.array([86,172,255])
mask = cv2.inRange(img_hsv, lower_epuck, upper_epuck)
img_epuck, contours_epuck, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

img_epuck = cv2.dilate(img_epuck, kernel, iterations = 7)
img_epuck, contours_epuck, hierarchy = cv2.findContours(img_epuck, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

center_epuck = [] # locating center of epuck
# finding centre of epuck contour
for c in contours_epuck:
    M = cv2.moments(c)
    center_epuck_X = int(M["m10"] / M["m00"])
    center_epuck_Y = int(M["m01"] / M["m00"])
    center_epuck.append([center_epuck_X,center_epuck_Y])

font = cv2.FONT_HERSHEY_SIMPLEX
textSize = cv2.getTextSize(char_direction, font, 1, 2)[0]
textX = int(center_epuck_X - textSize[0]/2)
textY = int(center_epuck_Y + textSize[1]/2)

cv2.circle(dst, (center_epuck[0][0], center_epuck[0][1]),25, (255,0,0), 3)
#cv2.drawContours(img_w, contours_wall, -1, (255,0,0), cv2.FILLED)
cv2.putText(dst, char_direction,(textX, textY), font, 1, (255,0,0), 2, cv2.LINE_AA)

img_wall_epuck = dst

plt.imshow(img_wall_epuck)
plt.show()
```

```python
# Import libraries
import cv2 
import numpy as np

img_destination = cv2.cvtColor(img_wall_epuck,cv2.COLOR_BGR2RGB)
img_correct_destination = np.zeros([500,900,3], np.uint8)
img_correct_destination.fill(255)

char_destination = 'x'
img_ladybug = cv2.imread(IMAGE_LADYBUG_FILE_NAME)
surf = cv2.xfeatures2d.SURF_create()

kp1, des1 = surf.detectAndCompute(dst, None) #query
kp2, des2 = surf.detectAndCompute(img_ladybug, None) #train

bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
matches = bf.match(des1, des2)
matches = sorted(matches, key = lambda x:x.distance)

list_kp1 = []
list_kp2 = []

for mat in matches[:20] :
    dst_idx = mat.queryIdx
    img_ladybug_idx = mat.trainIdx
    
    (x1,y1) = kp1[dst_idx].pt
    (x2,y2) = kp2[img_ladybug_idx].pt
    
    list_kp1.append((x1,y1))
    list_kp2.append((x2,y2))

x_destination_center = 0
y_destination_center = 0
#finding center of lady bug from list_kp1
for i in range(len(list_kp1)) :
    x = list_kp1[i][0]
    y = list_kp1[i][1]
    
    x_destination_center = x + x_destination_center
    y_destination_center = y + y_destination_center
    cv2.circle(img_correct_destination, (int(x),int(y)),1, (0,255,0), 3)
    
img_hsv = cv2.cvtColor(img_correct_destination,cv2.COLOR_BGR2HSV)

x_destination_center = int(x_destination_center/len(list_kp1))
y_destination_center = int(y_destination_center/len(list_kp1))

lower_destination = np.array([0,25,0])
upper_destination = np.array([179,255,255])
mask_destination = cv2.inRange(img_hsv, lower_destination, upper_destination)
img_correct_destination, contours_destination, hierarchy = cv2.findContours(mask_destination, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

img_correct_destination = cv2.dilate(img_correct_destination, kernel, iterations = 7)
img_correct_destination, contours_destination, hierarchy = cv2.findContours(img_correct_destination, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

index_destination = int(0)
largest_destination = len(contours_destination[0])

if len(contours_destination) > 0:
    for i in range(len(contours_destination)) :
        if len(contours_destination[i]) > largest_destination :
            largest_destination = len(contours_destination[i])
            index_destination = i

#print(index_destination)
#print(contours_destination[index_destination])

center_destination = []
for c in contours_destination :
    M = cv2.moments(c)
    center_destination_X = int(M["m10"] / M["m00"])
    center_destination_Y = int(M["m01"] / M["m00"])
    center_destination.append([center_destination_X,center_destination_Y])

font = cv2.FONT_HERSHEY_SIMPLEX
textSize = cv2.getTextSize(char_destination, font, 1, 2)[0]
textX = int(center_destination[index_destination][0] - textSize[0]/2)
textY = int(center_destination[index_destination][1] + textSize[1]/2)

cv2.putText(img_wall_epuck, char_direction,(textX, textY), font, 1, (0,0,255), 2, cv2.LINE_AA)
#img_epuck, contours_epuck, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

#print(center_dest)
#cv2.circle(img_destination, (x_destination_center,y_destination_center),25, (0,255,0), 3)
cv2.circle(img_destination, (center_destination[index_destination][0], center_destination[index_destination][1]),25, (0,255,0), 3)
cv2.putText(img_destination, char_destination,(textX, textY), font, 1, (0,255,0), 2, cv2.LINE_AA)
img_destination = cv2.cvtColor(img_destination,cv2.COLOR_BGR2RGB)
plt.imshow(img_destination)
plt.show()
```
